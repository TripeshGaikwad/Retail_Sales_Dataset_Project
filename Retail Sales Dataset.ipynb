```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

a = pd.read_csv("C:/Users/Sonu/OneDrive/Desktop/retail_sales_dataset.csv")

df = pd.DataFrame(a)
# print(df)

# print(a.info())
# a.describe()

# print(a.isnull().sum())

# print(a["Customer ID"].duplicated().sum())
# print(a.drop_duplicates("Customer ID"))





gb = df.groupby("Product Category").agg({"Total Amount":"sum"})
# print(gb)

#visualising
plt.bar(gb.index, gb["Total Amount"])
plt.title("Sales Based On Product Category")
plt.xticks(rotation = 90)
plt.xlabel("Product Category")
plt.ylabel("Total Sales Amount")
plt.show()


gb1 = df.groupby("Gender").agg({"Total Amount":"sum"})
# print(gb1)

plt.bar(gb1.index, gb1["Total Amount"])
plt.title("Sales Based On Gender")
plt.xticks(rotation = 90)
plt.xlabel("Product Category")
plt.ylabel("Total Sales Amount")
plt.show()


plt.violinplot(df["Age"], showmedians = True)
plt.title("Customer Segmentation Based On Age")
plt.ylabel("Customer Age")
plt.show()


plt.violinplot(df["Total Amount"], showmedians = True)
plt.title("Usual Purchase Amount") 
plt.ylabel("Purchase Amount")
plt.show()



# plt.violinplot(df["Price per Unit"], showmedians = True)
# plt.title("Product Pricing That People Usually Buy")
# plt.ylabel("Price Per Unit")
# plt.show()



# Create a violin plot with Seaborn
sns.violinplot(y="Price per Unit", data=df)
plt.title("Prices of Most Sold Products")
plt.ylabel("Price Per Unit")
plt.show()








# Your project is solid for inclusion in your resume, as it demonstrates key skills like data aggregation, visualization, and segmentation analysis. Here are a few suggestions to enhance it further:

# Exploratory Data Analysis (EDA):

# Provide insights or patterns you've noticed in the data.
# Include summary statistics, correlations, and distribution analysis.
# Data Cleaning:

# Mention any data cleaning steps you performed, like handling missing values or duplicates.
# Briefly describe the rationale behind these steps.
# Visualization Diversity:

# Consider adding different types of visualizations, like scatter plots for gender vs. age or sales trends over time.
# Use subplots to compare multiple categories in a single figure.
# Insights & Recommendations:

# Add a section summarizing key insights and potential business recommendations based on your analysis.
# Project Documentation:

# Ensure your code is well-commented and that you provide a clear project summary, including objectives and outcomes.
# These additions will make your project more comprehensive and impressive to potential employers.
```
